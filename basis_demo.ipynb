```python
from laguerre_basis import *

import matplotlib.pyplot as plt

k = np.arange(500)

#Each order is orthonormal:
first_basis = single_basis(k, 0.919, 1)
second_basis = single_basis(k, 0.919, 2)
print(first_basis.dot(first_basis.T))
print(second_basis.dot(second_basis.T))
print(first_basis.dot(second_basis.T))

basis_array = np.vstack((single_basis(k, 0.919, 0),single_basis(k, 0.919, 1),single_basis(k, 0.919, 2),single_basis(k, 0.919, 3),single_basis(k, 0.919, 4),single_basis(k, 0.919, 5),single_basis(k, 0.919, 6),single_basis(k, 0.919, 7),single_basis(k, 0.919, 8)))

eye_approx = basis_array.dot(basis_array.T)

#Bases are very nearly linearly independent
print(eye_approx)

#low l2 error
print(np.linalg.norm(eye_approx - np.eye(len(eye_approx))))

for idk in range(len(basis_array)):
    plt.plot(k, basis_array[idk][:], label="Laguerre Order {}".format(idk+1))

plt.legend()
plt.show()
```
